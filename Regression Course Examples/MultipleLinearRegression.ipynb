# Import Necessary Libraries
First, we need to import the required libraries.

```python
import numpy as np  # For numerical operations
import pandas as pd  # For handling datasets
import matplotlib.pyplot as plt  # For plotting
from sklearn.linear_model import LinearRegression  # Importing the Linear Regression model
```

# Prepare the Data
We create a dataset with three input variables (House Size, Number of Bedrooms, and House Age) and one target variable (House Price).

```python
# Define the independent variables (House Size, Bedrooms, House Age)
house_features = np.array([
    [50, 2, 10],  
    [75, 3, 5],  
    [100, 3, 8],  
    [125, 4, 4],  
    [150, 4, 2]
])
# Define the dependent variable (House Price in $)
house_prices = np.array([150000, 200000, 250000, 300000, 350000])
```

Understanding the dataset:

* X1 (House Size in m²)
* X2 (Number of Bedrooms)
* X3 (House Age in Years)
* Y (House Price in $)

# Train the Multiple Linear Regression Model
Now, we create and train the model using our dataset

```python
# Initialize the Linear Regression model
model = LinearRegression()
# Train the model using the dataset
model.fit(house_features, house_prices)
```

How it works?

.fit(X, Y) finds the best coefficients (m1, m2, m3, ...) and intercept (b) to minimize prediction errors.

# Retrieve the Learned Equation
After training, the model finds the best coefficients (m values) and intercept (b).

```python
# Get coefficients and intercept
m_values = model.coef_  # Slopes
b = model.intercept_  # Intercept
# Display the equation
print(f"Equation of the regression line: Y = {m_values[0]:.2f}X1 + {m_values[1]:.2f}X2 + {m_values[2]:.2f}X3 + {b:.2f}")
```

#  Make Predictions
Now that the model is trained, we can predict house prices based on new data.

```python
# New house data for prediction (Size, Bedrooms, Age)
new_house = np.array([[120, 3, 6]])
# Predict the price
predicted_price = model.predict(new_house)
# Display prediction
print(f"Predicted House Price: ${predicted_price[0]:.2f}")
```

# Visualizing the Relationship
Since we have three features, we can visualize the effect of one feature at a time while keeping others constant.

```python
# Plot House Size vs Price
plt.scatter(house_features[:, 0], house_prices, color='blue', label='Actual Prices')  # Actual prices
plt.scatter(new_house[:, 0], predicted_price, color='red', label='Predicted Price', marker='x', s=100)  # Predicted price
plt.xlabel("House Size (m²)")
plt.ylabel("House Price ($)")
plt.title("Effect of House Size on Price")
plt.legend()
plt.show()
```

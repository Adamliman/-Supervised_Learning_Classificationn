# Import Necessary Libraries
First, we need to import the required libraries.

```python
import numpy as np  # For numerical operations
import pandas as pd  # For handling datasets    
import matplotlib.pyplot as plt  # For plotting   
from sklearn.linear_model import LinearRegression  # Importing the Linear Regression model
```

# Prepare the Data
In this example, we have house sizes (in m²) as the input (X) and house prices (in $) as the output (Y).

```python
# Define the independent variable (House Size in m²)      
house_sizes = np.array([50, 75, 100, 125, 150]).reshape(-1, 1)  
# Define the dependent variable (House Price in $)  
house_prices = np.array([150000, 200000, 250000, 300000, 350000])  
```

Why Reshape (-1,1)?

Scikit-learn’s LinearRegression model expects input features (X) to be in a 2D array format, even if we have only one feature (house size).

Explanation of .reshape(-1, 1):

* -1 → Automatically determines the number of rows based on the input size.

* 1 → Ensures the data has one column, making it a 2D array suitable for training.

# Train the Linear Regression Model
Now, we create and train the model using the given data.

```python
# Initialize the Linear Regression model
model = LinearRegression()
# Train the model using the dataset
model.fit(house_sizes, house_prices)
```

```python
model.intercept_  # == b
```

```python
model.coef_   # == a
```

# Make Predictions
Now, we use the trained model to predict house prices.

```python
# Predict house prices based on input house sizes
predicted_prices = model.predict(house_sizes)
```

```python
predicted_prices
```

# Calculate the Difference Between Actual and Predicted Prices
To evaluate how well the model is performing, we compare the actual and predicted values.

```python
# Calculate the difference (error)
differences = house_prices - predicted_prices
# Print the differences
print("Differences between actual and predicted prices:", differences)
```

# Showing the results

```python
# Plot the data points
plt.scatter(house_sizes, house_prices, color='blue', label='Actual Prices')
# Plot the regression line
plt.plot(house_sizes, predicted_prices, color='red' , linewidth=2, label='Regression Line')
# Add labels and title
plt.xlabel('House Size (m²)')
plt.ylabel('House Price ($)')
plt.title('House Size vs. Price')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Exemple : plusieurs colonnes numériques
data = pd.DataFrame({
    'revenu': [1000, 1200, 1150, 980, 50000, 1100, 1080],
    'age': [25, 27, 30, 26, 45, 29, 500],
    'depenses': [200, 210, 190, 205, 4000, 195, 198]
})

# Boucle sur chaque colonne numérique
for col in data.select_dtypes(include='number').columns:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    # On supprime les outliers pour cette colonne
    data = data[(data[col] >= lower) & (data[col] <= upper)]

print("Données nettoyées :")
print(data)
```

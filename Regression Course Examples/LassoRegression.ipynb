# Import Necessary Libraries
First, we need to import the required libraries.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
```

# Create and Prepare the Dataset

```python
# Example dataset (house size, number of bedrooms vs price)
X = np.array([[500, 3], [600, 4], [700, 4], [800, 5], [900, 5], [1000, 6], [1100, 6], [1200, 7]])
Y = np.array([150000, 180000, 210000, 240000, 270000, 300000, 330000, 360000])
```

# Splitting the Data into Training and Testing Sets

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

# Training the Lasso Regression Model


```python
lasso_model = Lasso(alpha=1.0)  # λ = 1.0 (Regularization strength)
lasso_model.fit(X_train, Y_train)
```

# Make Predictions

```python
Y_pred = lasso_model.predict(X_test)
print("Predicted Prices:", Y_pred)
```

# Analyzing the Feature Selection Effect

```python
print("Lasso Coefficients:", lasso_model.coef_)
print("Intercept:", lasso_model.intercept_)
```

Key Observation:

* Some coefficients may be exactly zero, meaning those features were removed.
* This shows how Lasso automatically selects the most important features.

# Visualizing Lasso Regression

```python
plt.scatter(X[:, 0], Y, color='blue', label='Actual Prices')
plt.plot(X[:, 0], lasso_model.predict(X), color='red', linewidth=2, label="Lasso Regression")
plt.xlabel("House Size (sq ft)")
plt.ylabel("House Price ($)")
plt.title("House Price Prediction using Lasso Regression")
plt.legend()
plt.show()
```

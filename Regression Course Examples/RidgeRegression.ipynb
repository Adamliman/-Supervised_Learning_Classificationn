# Import Necessary Libraries
First, we need to import the required libraries.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
```

# Create & Prepare the Data

```python
# Example dataset (house size vs price)
X = np.array([[500], [600], [700], [800], [900], [1000], [1100], [1200]])
Y = np.array([150000, 180000, 210000, 240000, 270000, 300000, 330000, 360000])
```

# Splitting the Data into Training and Testing Sets

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

# Training the Ridge Regression Model

```python
ridge_model = Ridge(alpha=1.0)  # λ = 1.0 (Regularization strength)
ridge_model.fit(X_train, Y_train)
```

Explanation

* The alpha parameter (λ) controls the degree of regularization.
* The model optimizes coefficients while minimizing overfitting.

# Make Predictions
Now, we use the trained model to predict house prices.

```python
Y_pred = ridge_model.predict(X_test)
print("Predicted Prices:", Y_pred)
```

# Visualizing the Ridge Regression Model

```python
plt.scatter(X, Y, color='blue', label='Actual Prices')
plt.plot(X, ridge_model.predict(X), color='red', linewidth=2, label="Ridge Regression")
plt.xlabel("House Size (sq ft)")
plt.ylabel("House Price ($)")
plt.title("House Price Prediction using Ridge Regression")
plt.legend()
plt.show()
```

# Comparing Ridge and Linear Regression
To analyze the impact of regularization, we compare Ridge with standard Linear Regression.

```python
from sklearn.linear_model import LinearRegression
# Train a simple Linear Regression model
linear_model = LinearRegression()
linear_model.fit(X_train, Y_train)
# Compare coefficients
print("Linear Regression Coefficients:", linear_model.coef_)
print("Ridge Regression Coefficients:", ridge_model.coef_)
```

Key Observations

* Ridge Regression shrinks the coefficients, preventing overfitting.
* Linear Regression does not penalize large coefficients, which can lead to instability.

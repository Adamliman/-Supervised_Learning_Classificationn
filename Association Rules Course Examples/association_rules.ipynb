## 🛒 Apriori Situation: Market Basket Analysis

Imagine you're a **Machine Learning Engineer** working for a supermarket. Your goal? 🧠 Discover valuable patterns in customer purchases to help improve business decisions.

Your employer hands you a dataset containing customer transactions that looks like this:

```python
dataset = [ ['Milk', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'], ['Dill', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'], ['Milk', 'Apple', 'Kidney Beans', 'Eggs'], ['Milk', 'Unicorn', 'Corn', 'Kidney Beans', 'Yogurt'], ['Corn', 'Onion', 'Onion', 'Kidney Beans', 'Ice cream', 'Eggs'] ]
```

🧾 Transaction Format: Each inner list in our dataset represents a **transaction** — a purchase made by a customer

## 🛠️ Apriori Preparation

Before applying the **Apriori algorithm**, we need to **preprocess our dataset** properly.

### 🔄 One-Hot Encoding

The Apriori function requires a dataset where:
- Each **row** represents a transaction
- Each **column** represents a unique item
- Each **cell** contains either **True** or **False** indicating the presence or absence of that item in the transaction

This transformation is known as **one-hot encoding**.

### 🧰 Using `TransactionEncoder`

To achieve this, we use the `TransactionEncoder` class from the **`mlxtend`** library. It:
- Takes a list of transactions as input
- Returns a NumPy array with boolean values (True/False)
- Not 0/1 — but functionally the same idea

```python
import mlxtend
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
te=TransactionEncoder()
te_ary=te.fit(dataset).transform(dataset)    #Apply one-hot-encoding on our dataset
df=pd.DataFrame(te_ary, columns=te.columns_)  #Creating a new DataFrame from our Numpy array
df
```

## 📊 Support Code – Filtering Itemsets by Support

Now that the dataset has been transformed using one-hot encoding, we can move on to **extracting frequent itemsets** using the **Apriori algorithm**.

🎯 Goal: We want to find itemsets that appear in **at least 60%** of all transactions — this is called setting the **minimum support threshold**.

```python
from mlxtend.frequent_patterns import apriori
apriori(df, min_support=0.6)
```

### Support with column names

```python
frequent_itemsets=apriori(df, min_support=0.6, use_colnames=True) #Instead of column indices we can use column names.
frequent_itemsets
```

## ✅ Confidence-Based Rule Extraction

Once we have our frequent itemsets, we can extract **association rules** using other interesting metrics like **confidence** and **lift**.

```python
from mlxtend.frequent_patterns import association_rules
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.7) # associate items
```

## 🚀 Lift-Based Rule Extraction

In addition to **confidence**, we can also use **lift** as a metric to extract more meaningful association rules.

```python
from mlxtend.frequent_patterns import association_rules
association_rules(frequent_itemsets,metric="lift",min_threshold=1.25)
```

# Association Rule Mining Metrics

## 📌 Basic Components
- **Antecedents**: Items on the **left side** of the rule (e.g., `{Milk} → {Bread}`, where "Milk" is the antecedent).
- **Consequents**: Items on the **right side** of the rule (e.g., "Bread" in the above example).

## 📊 Key Metrics

| **Metric**            | **Definition** | **Formula / Explanation** |
|----------------------|--------------|---------------------------|
| **Antecedent Support** | Probability of antecedent occurring | `Support(A) = Transactions(A) / Total Transactions` |
| **Consequent Support** | Probability of consequent occurring | `Support(B) = Transactions(B) / Total Transactions` |
| **Support**          | How often both antecedent & consequent appear together | `Support(A → B) = Transactions(A, B) / Total Transactions` |
| **Confidence**       | Likelihood of B appearing given that A appears | `Confidence(A → B) = Support(A, B) / Support(A)` |
| **Lift**            | Strength of association compared to random occurrence | `Lift(A → B) = Confidence(A → B) / Support(B)` |
| **Representativity** | Indicates how well a rule represents the dataset | Measures how frequent a rule appears |
| **Leverage**        | Difference between actual & expected support if A and B were independent | `Leverage(A → B) = Support(A, B) - (Support(A) * Support(B))` |
| **Conviction**      | Measures dependency between antecedent & consequent | `Conviction(A → B) = (1 - Support(B)) / (1 - Confidence(A → B))` |
| **Zhang's Metric**  | Balances impact of both antecedent and consequent | Measures how interesting the rule is |
| **Jaccard**         | Similarity between itemsets | `Jaccard(A, B) = Support(A, B) / (Support(A) + Support(B) - Support(A, B))` |
| **Certainty**       | Reliability of rule based on confidence and dataset | Helps gauge strength of rule |
| **Kulczynski**      | Balances confidence in **both directions** | `Kulczynski(A → B) = (Confidence(A → B) + Confidence(B → A)) / 2` |

## 🎯 Summary
Association rule mining helps uncover hidden patterns in datasets by evaluating relationships between items using **support, confidence, lift**, and other statistical metrics.
